### Listas de Python y sus limitaciones

La [listas](https://docs.python.org/es/3/tutorial/introduction.html#lists) son estructuras de datos muy versátiles consistentes en secuencias mutables de objetos indexados, con gran flexibilidad para acceder a sus conjenidos mediante índices o rebanadas ([slices](https://docs.python.org/es/3/glossary.html#term-slice)). Aunque los objetos que las componen pueden ser heterogéneos, los más frecuente es que sean de la misma clase, por ejemplo, una lista de números enteros. 

Las listas se construyen delimitando entre corchetes una secuencia de objetos separados por comas.

Veamos primeramente un a lista de objetos heterogéneos,compuesta por números enteros, reales, booleanos e, incluso, por otras secuencias, tales como cadenas y tuplas.

```python
lista_a = ['El número entero ', 5, 'unas coordenadas: ', (2,3), 
           'un real: ', 7.3, '¿Será cierto? ', True]
print(lista_a)
print(type(lista_a))
print('longitud: ', len(lista_a))
```

Podemos acceder a elementos de la lista (igualmente se haría con otras secuencias, como cadanas y tuplas) por:
- Índice, simplemente indicándolo dentro de un corchete. El primer elemento de la lista tiene índice \[0\], de modo que el último tiene de índice la \[*longitud* - 1\]. Es interesante advertir que también podemos numerar los índices desde el final de la lista, utilizando números negativos. De este modo, el último elemento tiene índice \[-1\] y el primero, tienen índice \[-*longitud*\].
- Rebanadas ([slices](https://docs.python.org/es/3/glossary.html#term-slice)), que refiere una porción de una secuencia mediante la notación \[*inicio*:*final*+1:*salto*\] para indicar el rango de índices, que pueden ser, como ya se ha indicado, positivos o negativos. El *salto* también puede ser positivo o negativo, según progrese a lo largo de la lista o en sentido contrario. Si no se indican, los componentes de la rebanada toman valores por defecto \[0:*longitud*:1\].

Esta forma de acceso a los elementos que estamos viendo con las [listas](https://docs.python.org/es/3/tutorial/introduction.html#lists) es **válida también para** [cadenas](https://docs.python.org/es/3/tutorial/introduction.html#strings) y para [tuplas](https://docs.python.org/es/3/tutorial/datastructures.html#tuples-and-sequences), y es la base para referirise a los componentes de otras estructuras de datos que veremos. Es por ello muy importante entenderlo bien.

Veamos, en primer lugar, algunos ejemplos de acceso mediante índice:

```python
print(lista_a[0]) #El índice comienza en 0
print(lista_a[-1]) # El último elemento
print(lista_a[4], lista_a[-3]) # elementos intermedios
```

Y veamos ahora algunos ejemplos de acceso con [rodajas (*slices*)](https://docs.python.org/es/3/glossary.html#term-slice). Es muy importante que intentes entenderlo. Puedes hacer cambios sobre el *notebook*, o abrirte uno nuevo, para experimentar:

```python
print(lista_a[:])
print(lista_a[0:8]) # Comienza en 0 y temina en longitud - 1
print(lista_a[::-1])
print('\n')
print(lista_a[2:])
print(lista_a[-2:])
print('\n')
print(lista_a[2:4])
print(lista_a[2:6:2])
print('\n')
print(lista_a[::2])
print(lista_a[::-2])
```

Hemos comentado comentado que las listas son secuencias de objetos que pueden ser heterogéneos, y que podemos acceder a los contenidos de la cadenas y las secuencias con índices y rodajas como hacemos con las listas.

¿Entiendes qué estamos haciendo con el código que sigue?

```python
# Volvemos a definir lista_a por si se ha variado experimentando con el notebook
lista_a = ['El número entero ', 5, 'unas coordenadas: ', (2,3), 
           'un real: ', 7.3, '¿Será cierto? ', True]
print('lista_a: ', lista_a, '\n')

print(lista_a[0][3:9])
print(lista_a[3][1])
```

Hemos que comentado **las listas son secuencias mutables**, esto es, que las podemos modificar una vez creado en memoria. Esto no podría hacerse **ni con cadenas ni con tuplas, pues son inmutables**. 

Las listas son objetos, con métodos que permiten editarlas. Puedes ver los més frecuentes en la [extensión sobre listas](https://docs.python.org/es/3/tutorial/datastructures.html#more-on-lists) del tutorial de Python.

Veamos algunos ejemplos. Puedes experimentar sobre el *notebook* o sobre uno nuevo. 

```python
# Volvemos a definir lista_a por si se ha variado experimentando con el notebook
lista_a = ['El número entero ', 5, 'unas coordenadas: ', (2,3), 
           'un real: ', 7.3, '¿Será cierto? ', True]
print('lista_a: ', lista_a)

# Ahora la modificamos
print('\nModificaciones de la lista_a:')
lista_a[-1] = False # Cambiamos de True a False
print(lista_a)
lista_a[::2] = ['Uno', 'Dos', 'Tres', 'Cuatro']
print(lista_a)
```

El nombre de un objeto es una referencia al mismo en memoria, no el objeto en sí mismo. Esto hace que puedan encontrarse efectos tal vez inesperados, que hay que tratar con cuidado.

De hecho, hay en Python dos tipos de **copias de objetos**:
- Superficiales: solo se crea un nuevo nombre para acceder al objeto, de modo que ambos nombres refieren al mismo objeto en memoria.
- Profundas: creamos un nuevo objeto en memoria y copiamos, uno a uno, los elementos del anterior. Ahora tendremos dos objetos en memoria.

Muchas operaciones sobre objetos mutables son superficiales.

Reflexiona qué está pasado en este ejemplo:

```python
lista_b = lista_a

print(lista_a)
print(lista_b)

print('\nModificamos lista_b. ¿Eso afecta a lista_a?')
lista_b.append('Cambio')
print('lista_a: ', lista_a)
print('lista_b: ', lista_b)
```

#### Listas anidadas

Como hemos visto, las listas son secuencias de objetos. Si hacemos listas cuyos elementos son, a su vez, listas, obtnemos listas anidades. Esto resulta muy útil para connstruir *arrays* multidimensionales.

Una **imagen digital** puede representarse con un **array bidimensional** de números enteros. Construyamos un array bidimensional muy sencillo de 6 filas x 8 culumnas para representar una imagen cuya mitad superior sea brillante y la inferior sea negra: 

Adviértase lo complicado que resulta crear y poblar el *array*, que vamos haciendo crecer sucesivamente en memoria con nuevos valores. Esto resulta muy ineficiente.

```python
array_2D = []
for i in range(6):
    array_2D.append([])
    k = 100 if i <=2 else 0
    for j in range(8):
        array_2D[i].append(k)

print(type(array_2D))
print('Nº filas: ', len(array_2D))
print('Nº columnas: ', len(array_2D[0])) # Todas las filas son iguales
display(array_2D)
```

Modifiquemos ahora el *array* para incluir un rectángulo interior con brillo superior:

```python
for i in [2, 3]:
    for j in [2, 3, 4, 5]:
        array_2D[i][j]= 255
        
display(array_2D)
```

Hemos creado un array bidimensional que representa una imagen y que podemos interpretar, en este caso, al ser muy pequeña. Sin embargo, su creación y la operación de sus elementos resulta enormemente tediosa y computacionalmente ineficiente. Tampoco tenemos una manera sencilla de visualizar el *array* como una imagen, ni siquiera utilizando PILLOW.

#### Comprensiones de listas

Las listas pueden crearse y accederse a sus elementos mediante un constructo denominado comprensión de listas (también hay comprensiones de otras secuencias), que es más cómodo y eficiente.

Creemos primero un *array* 1D, que inicializamos a cero y del que, a continuación, modificamos valores intermedios

```python
array_1D = []
array_1D = [0  for j in range(8)]
display(array_1D)
array_1D[2:6] = [255]*4
display(array_1D)
array_1D[4] = 100
display(array_1D)
```

Podemos también anidar las comprensiones de listas para crear arrays multidimensionales, por ejemplo, un array bidimensional como el que se muestra a continuación.

El **operador ternario condicional** resulta  muy útil para seleccionar uno entre dos valores posibles.

```python
array_2D_bis = []
array_2D_bis = [[0 if j<4 else 100  for j in range(8)] for i in range(6)]
display(array_2D_bis)
```

#### Una breve introducción a Numpy

Como ya hemos visto, **Numpy** introduce una nueva clase de objetos *ndarray* para trabajar de forma eficiente con *arrays* multidimensionales y operaciones matemáticas.

Comencemos viendo cómo se indexan los **arrays unidimensionales**. 

* Como las listas
  * Se indexan a partir de 0, con índices enteros y conforme a las mismas reglas que vimos para las *listas* de Python
  * Igual que hacíamos con las listas de Python, podemos extraer porciones de *arrays unidimensionales* utilizando rebanadas o *slices*
  * Son **mutables**
* De forma específica para los *ndarrays*, podemos utilizar **fancy indexing**, que consiste en indexar el *array* mediante listas o *ndarrays*, lo que, como veremos, añade una gran flexibilidad
  * Con los índices enteros, definidos en una lista o en otro *ndarray*, que no tienen que guardar un orden y pueden estar repetidos.
  * Con una máscara booleana de igual longitud que el ndarray
* También, a diferencia de las listas, hay multitud de formas de crear *ndarrays* y de operar con ellos.

El indexado de Numpy puede resultar a veces confuso, de forma que es necesario recurrir a la [documentación de indexado](https://numpy.org/doc/stable/user/basics.indexing.html#indexing-on-ndarrays) con frecuencia. El [capítulo de indexado](https://scipy-lectures.org/intro/numpy/array_object.html#indexing-and-slicing) en las SciPy Lecture Notes es muy ilustrativo.

Vamos primeramente a crear algunos arrays unidimensionales  de ejemplo. En cada caso, los componentes de los arrays son del mismo  tipo, para reservar la memoria desde un principio. No tienen que ser necesariamente numéricos. De hecho, crearemos también un array de cadenas de caracteres, cuyos componentes tendrán todos el tamaño de la mayor cadena utilizada en la inicialización (recordar que todos los objetos del array son igual tipo y tamaño para optimizar su manejo.

```python
nd_array_1D_1 = np.array([1, 2, 3, 4])
display(type(nd_array_1D_1))
display(nd_array_1D_1.dtype)
display(nd_array_1D_1.shape)
display(nd_array_1D_1)
```

```python
nd_array_1D_2 = np.array([1, 2, 3, 4], dtype=float)
display(type(nd_array_1D_2))
display(nd_array_1D_2.dtype)
display(nd_array_1D_2.shape)
display(nd_array_1D_2)
```

```python
nd_array_1D_3 = np.array([1+0j, 0+1j, -1+0j, 0-1j])
display(type(nd_array_1D_3))
display(nd_array_1D_3.dtype)
display(nd_array_1D_3.shape)
display(nd_array_1D_3)
```

Veamos ahora cómo utilizar el indexado y las rabanadas (*slices*) de las listas con los *ndarrays*. Es inmediato, con tal de que sepas hacerlo con las listas.

```python
print('nd_array_1D_1 = ', nd_array_1D_1, ' --> nd_array_1D_1[1] = ', nd_array_1D_1[1])
print('nd_array_1D_2 = ', nd_array_1D_2, ' --> nd_array_1D_2[-1] = ', nd_array_1D_2[-1])
print('nd_array_1D_3 = ', nd_array_1D_3, ' --> nd_array_1D_3[1:3] = ', nd_array_1D_3[1:3])
print('nd_array_1D_4 = ', nd_array_1D_4, ' --> nd_array_1D_4[::2] = ', nd_array_1D_4[::2])
```

El [fancy indexing](https://scipy-lectures.org/intro/numpy/array_object.html#fancy-indexing), sin embargo, es un método de indexado propio de Numpy, que no existe en las listas. Tiene dos modalidades:
- Indexar con un **array de booleanos**, que tenga exactamente la **misma longitud que el array que se indexa**. Se seleccionan los elementos correspondientes a los valores *True*
- Indexar con una **lista o o con un array de enteros**, de una **longitud arbitraria** y se extraen en un nuevo *ndarray* los valores correspondientes a los índices que se refieren. 

Utilizaremos el *ndarray booleano* que creamos antes para hacer algunos ejemplos.

Es este un buen momento para hablar de las [copias y las vistas de los ndarrays](https://scipy-lectures.org/intro/numpy/array_object.html#copies-and-views), conceptos intrínsicamente vinculados a la naturaleza **mutable** de los *ndarrays* de Numpy.

Si *copiamos* un array simplemente asignándolo a otra variable, el array sigue siendo el mismo, tan solo tenemos dos nombres distintos para referirlo en memoria, tal y como ya vimos con las listas. Podemos comprobarlo invocando el **identificador de objeto** con ambos nombres y viendo que son **idénticos**.

¿Qué pasa cuando generamos un array con un método o función? Si la operación se refiere a algún tipo de  reodenación, con frecuencia (pero no siempre, depende de la implementación del método o función) se mantienen los mismos valores en memoria, de modo que el resultado consiste en una estructura (un nuevo objeto con su propio identificador)  que hace referencia a los valores anteriores, denominada **vista**, que son compartidos en memoria por la vista y el *ndarray* original. Si modificamos el *ndarray* original cambia la vista y viceversa, modificando la vista también lo hacemos con el *ndarray* original.

Las vistas son muy eficientes computacionanalmente, especialmente en operaciones de reordenación, indexado y *rebanado*, pero pueden ocasionarse efectos inesperados si no se entiende bien en qué consiste, modificándose inadvertidamente *ndarrays* al actuar en sus vistas.

A veces necesitamos hacer **copias profundas** que realmente creen en memoria los nuevos valores y nos permitan trabajar independientemente con las estructuras de datos copiadas. Es a lo que simplemente se suele llamar **copias**.

Utilizaremos nuestro *ndarray* de cadenas para ilustrar estos conceptos. Si haces experimentos, ten en cuenta el tamaño máximo que pueden tener las cadenas.

```python
# SIMPLE ASIGNACIÓN DE NOMBRE
print('nd_array_1D_4 = ', nd_array_1D_4)
print('id(nd_array_1D_4) = ', id(nd_array_1D_4))
nd_array_1D_4_bis = nd_array_1D_4
print('id(nd_array_1D_4_bis) = ', id(nd_array_1D_4_bis))
nd_array_1D_4_bis[1] = 'BIS' 
print('nd_array_1D_4_bis = ', nd_array_1D_4_bis)
print('¿Cambia nd_array_1D_4?')
print('nd_array_1D_4 = ', nd_array_1D_4)
np.may_share_memory(nd_array_1D_4, nd_array_1D_4_bis)
```

```python
# Reconstrucción
nd_array_1D_4[1] = 'adiós'
print('nd_array_1D_4 = ', nd_array_1D_4)
print('id(nd_array_1D_4) = ', id(nd_array_1D_4))
print('id(nd_array_1D_4_bis) = ', id(nd_array_1D_4_bis))
print('nd_array_1D_4_bis = ', nd_array_1D_4_bis)
```

```python
# VISTA
print('nd_array_1D_4 = ', nd_array_1D_4)
print('id(nd_array_1D_4) = ', id(nd_array_1D_4))
# Creamos vista
nd_array_1D_4_vista = nd_array_1D_4[::-1]
print('id(nd_array_1D_4_vista) = ', id(nd_array_1D_4_vista))
nd_array_1D_4_vista[1] = 'VISTA'
print('nd_array_1D_4_vista = ', nd_array_1D_4_vista)
print('¿Cambia nd_array_1D_4?')
print(np.may_share_memory(nd_array_1D_4, nd_array_1D_4_vista))
print('nd_array_1D_4 = ', nd_array_1D_4)
```

```python
# Reconstrucción
nd_array_1D_4_vista[1] = 'perfecto'
print('nd_array_1D_4 = ', nd_array_1D_4)
print('id(nd_array_1D_4) = ', id(nd_array_1D_4))
print('nd_array_1D_4_vista = ', nd_array_1D_4_vista)
print('id(nd_array_1D_4_vista) = ', id(nd_array_1D_4_vista))
```

```python
# COPIA
print('nd_array_1D_4 = ', nd_array_1D_4)
print('id(nd_array_1D_4) = ', id(nd_array_1D_4))
# Creamos copia
nd_array_1D_4_copia = nd_array_1D_4[::-1].copy()
print('id(nd_array_1D_4_copia) = ', id(nd_array_1D_4_copia))
nd_array_1D_4_copia[1] = 'COPIA'
print('nd_array_1D_4_copia = ', nd_array_1D_4_copia)
print('¿Cambia nd_array_1D_4?')
print(np.may_share_memory(nd_array_1D_4, nd_array_1D_4_copia))
print('nd_array_1D_4 = ', nd_array_1D_4)
```

Para [construir ndarrays](https://numpy.org/doc/stable/user/basics.creation.html) ya hemos visto que podemos invocar la función de factoría *numpy.array()* sobre una secuencia con objetos de la misma clase, y que podemos forzar el tipo de los objetos del nuevo *ndarray*.

Hay otras muchas formas de [crear nuevos ndarrays](https://numpy.org/doc/stable/user/basics.creation.html). Veremos ahora algunos ejemplos comunes:
- np.zeros(), np.ones() para crear arrays con elementos idénticos (unos o ceros)
- np.indices(), np.arange(), np.linspace() para crear rrays con rangos de valores

Podemos también hacer muy fácilmente [operaciones sobre los ndarrays y sus elementos](https://scipy-lectures.org/intro/numpy/operations.html).

Empecemos viendo *np.zeros(shape)*, *np.ones(shape)* y algunas operaciones básicas. *shape* es una tupla con la dimensiones del *ndarray* que, al ser, de momento, unidimensionales solo tienen una componente con la longitud: *shape = (long, )*. Adviértase que ponemos una coma en la tupla aunque no tenga segunda componente, pues de otro modo podría confundirse con un número aislado. 

```python
ceros = np.zeros((10,))
unos = np.ones((10,))
indx = np.indices((10,))
print('ceros = ', ceros, '\n', 'unos =', unos, '\n', 'indx = ', indx, '\n')

print('ceros + 5 =', ceros +5)
print('unos + indx =', unos + indx)
print('ceros*unos =', ceros*unos)
print('indx^2 = ', indx**2)
```

Podemos también efectuar fácilmente operaciones lógicas sobre los *ndarrays*, para obtener nuevos arrays que, a su vez, podemos utilizar para hacer *fancy indexing*.

Veamos por ejemplo como extraer los números pares de *indx^2*:

```python
display((indx**2) % 2 == 0)
print((indx**2)[indx**2 % 2 == 0])
```

Veamos cómo [crear ndarrays unidimensionales con rango de valores](https://numpy.org/doc/stable/user/basics.creation.html#d-array-creation-functions):
- np.arange(inicio, final, salto) es el equivalente a la función range vista en las listas. Construye un array cuyos elementos comienzan en *inicio* y van avanzando incrementando *satto*, hasta llegar a final, que NO se incluye en el ndarray.Esta función es muy útil con enteros, pero pueda dar lugar a que el último elemento se incluya o no dependiendo de la precisión de la operación cuando se trabaja con reales, de modo que no podemos estar seguros de antemano del número de elementos que resultarán (cuando se trabaja en punto flotante).
- np.linspace(inicio, final, número) construye un array con *número* de elementos equiespaciados entre inicio y final, sin incluir al último. Ahora sí vamos a estar seguros de que el *ndarray* resultante tenga *número* de elementos. Existe una variante en el que el espaciado es logarítmico: np.logspace()


```python
grados = np.arange(0,390, 30)
print(grados.dtype, grados.shape)
display(grados)
```

```python
radianes = np.linspace(0, 2*np.pi, 13)
print(radianes.dtype, radianes.shape)
display(radianes)
```

Hemos creado un array de la circunferencia completa en grados (utilizando np.arange) y otro en radianes (utilizando np.linspace), con idéntico número de elementos, y les calculamos la función seno a cada uno para obtener nuevos *ndarray*.

Teóricamente, el senos calculado en grados o en radianes debe dar un resultado idéntico. Pero al trabajar con precisión finita, puede haber una pequeña diferencia como, de hecho, podemos comprobar haciendo la comparación aunque probablemente no podremos observar pues la diferencia estará en un decimal muy pequeño (puede variar dependiendo de la arquitectura) 

```python
sen_grad = np.sin(grados*np.pi/180)
sen_rad = np.sin(radianes)

print('sen_grad = ', sen_grad)
print('sen_rad = ', sen_rad)
print('\nElementos "diferentes" por precisión finita:')
print('sen_grad == sen_rad -->', sen_grad == sen_rad)
print('sen_grad -->', sen_grad[sen_grad != sen_rad])
print('sen_rad -->', sen_rad[sen_grad != sen_rad])
```

Veamos ahora **ndarrays multidimensionales**. Por ejemplo, una imagen en niveles de gris la representaremos con un *ndarray bidimensional* mientras que, si es en color, utilizaremos un *ndarray tridimensional*.

Podemos hacer [operaciones lógicas](https://numpy.org/doc/stable/reference/routines.logic.html) sobre los elementos de los arrays, y generar máscaras que nos permitan **clasificar** y extraer los elementos según sean sus valores.

Python tiene una clase booleana con dos elementos, True y False, que pueden evaluarse numéricamente (True -> 1 y False -> 0). Sin embargo, debemos tener cuidado en las operaciones, para que los tipos sean compatibles. Si bien el intérprete puede hacer algunas conversiones (ej 1 + False), es conveniente ser coherentes con los tipos al aplicar las operaciones para evitar resultados inesperados.

El método [ndarray.astype()](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.astype.html?highlight=numpy%20astype#numpy.ndarray.astype) permite realizar cambios de tipo en los elementos de un array. Hay que ser muy cuidadosos pues, al tener diferentes tipos precisiones distintas, pueden producirse resultados y conversiones inesperados, pues Numpy intentará realizar la conversión transformando valores conforme a sus reglas de conversión. 

Por ejemplo, en el ejemplo siguiente se convierte un array de booleanos (la máscara) a tipo byte, con la conversión ya mencionada (True -> 1 y False -> 0).

Para visualizar imágenes sobre Jupyter será necesario trabajar con los [modos L y RGB](https://pillow.readthedocs.io/en/stable/handbook/concepts.html#modes) de PILLOW, lo que puede hacer necesaria una conversión (y un reescalado, para que los nímeros estén entre 0 y 255) de los ndarrays al tipo *unsigned byte*.

```python
mask = (np.logical_and(np_img >25, np_img<125)).astype('ubyte') 

display(Image.fromarray(mask*np_img))
```

### Un ejemplo con una imagen real

Vamos a aplicar lo que hemos visto a una imagen RGB real que, primeramente, leemos usando PILLOW.

```python
# Imagen cogida de https://pixabay.com/es/photos/perro-jack-russell-terrier-terrier-7417233/
file = 'images/perro.jpg'
img  = Image.open(file)
print(type(img))
print('formato:', img.format)
print('(ancho, alto) =', img.size)
print('modo: ', img.mode)
display(img)
```

Convertimos la imagen a un *ndarray* de Numpy:

```python
img_np = np.array(img)
print(type(img_np))
print(img_np.dtype)
print('(alto, ancho, bandas) =', img_np.shape)
```

Creamos tres arrays con los canales rojo, verde y azul. Aunque podríamos representarlos en niveles de gris, vamos a representarlos cada uno en color. Para que la imagen original y los canales no interfieran entre sí, vamos a hacer **copias profundas** con el método *copy()*  

```python
img_R = np.zeros(img_np.shape, dtype ='ubyte')
img_R[:,:,0] = img_np[:,:,0].copy()
display(Image.fromarray(img_R))
```

```python
img_G = np.zeros(img_np.shape, dtype ='ubyte')
img_G[:,:,1] = img_np[:,:,1].copy()
display(Image.fromarray(img_G))
```

```python
img_B = np.zeros(img_np.shape, dtype ='ubyte')
img_B[:,:,2] = img_np[:,:,2].copy()
display(Image.fromarray(img_B))
```

Ahora vamos a generar unas máscaras para cada color, que indicarán qué color tiene intensidad máxima para cada píxel, y las integraremos en un *ndarray* booleano de tres canales, uno para cada máscara.

En la visualización de la máscara veremos cada pixel en rojo, verde o azul según sea el valor superior de intensidad de los tres canales (por ejemplo, rojo quiere decir que el píxel en ese canal tiene mayor intensidad que en azul o verde). Los píxels en negro corresponden a los empates entre canales.

```python
mask = np.zeros(img_np.shape, dtype ='bool')
print(type(mask))
print(mask.dtype)
print('(alto, ancho, bandas) =', mask.shape)

mask[:,:,0] = np.logical_and(img_np[:,:,0] > img_np[:,:,1], img_np[:,:,0] > img_np[:,:,2])
mask[:,:,1] = np.logical_and(img_np[:,:,1] > img_np[:,:,0], img_np[:,:,1] > img_np[:,:,2])
mask[:,:,2] = np.logical_and(img_np[:,:,2] > img_np[:,:,0], img_np[:,:,2] > img_np[:,:,1])
```

```python
display(Image.fromarray(255*mask.astype('ubyte')))
```
