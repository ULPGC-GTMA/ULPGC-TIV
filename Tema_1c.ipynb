### La interfaz de programación (API) imperativa de Matplotlib

*matplotlib.pyplot* es una colección de funciones que hacen que Matplotlib funcione de forma parecida a MATLAB en las representaciones gráficas. Cada función de *pyplot* realiza algún cambio en una figura: por ejemplo, crea una figura, crea un área de dibujo en una figura, dibuja algunas líneas en un área de dibujo, decora el dibujo con etiquetas, etc. Internamente, *pyplot* actualiza y almacena el estado de todos los objetos de la figura.

*pyplot* está diseñado principalmente para gráficos interactivos y para casos simples de generación de gráficos programáticos. Es muy simple de utilizar, pero no es la opción recomendada salvo en tales casos sencillos. Para aplicaciones más complejas se recomienda la API orientada a objetos, que también se introduce brevemente después.

En la documentación oficial hay disponible un [Tutorial de pyplot](https://matplotlib.org/stable/tutorials/introductory/pyplot.html) y [ejemplos](https://matplotlib.org/stable/gallery/pyplots/index.html#pyplots-examples).


Comenzaremos importando *pyplot* y *numpy* para ver un primer ejemplo muy sencillo. Si bien habitualmente los casos de uso de *Matplotlib* utilizan *ndarrays* de Numpy, es posible utilizar directamente listas de Python (que internamente *Matplotlib* convierte en *ndarrays*).  

La función clave es *plot*, que nos permite dibujar una función de una variable independiente. Podemos utilizar distintos tipos de líneas, de marcadores y colores. También podemos superponer varias gráficas. La sintaxis de esta función es muy versátil, y, para saber más, puede consultarse la [referencia de pyplot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot). 

Podemos acompañar los ejes con texto y poner un título a la gráfica con las funciones *xlabel()*, *ylabel()* y *title()*. Además, la función *text()* nos permite incluir texto en posiciones arbitrarias (las coordenadas en que se inserta coinciden con las de los ejes). Las propiedades del texto (fuente, tamaño, color, etc) pueden variarse introduciendo las propiedades que correspondan en la llamada a la función. Podemos también incluir texto en [Latex](https://matplotlib.org/stable/tutorials/text/mathtext.html) para escribir expresiones matemáticas. La guía de [texto en plots de Matlib](https://matplotlib.org/stable/tutorials/text/text_intro.html) da más información. 

Es muy útil también anotar las gráficas, para destacar señalando lo que corresponda, con la función *annotate()*. Puedes ver más sobre cómo *anotar* gráficas en la [guía de anotaciones de Matplotlib](https://matplotlib.org/stable/tutorials/text/annotations.html#id1). Pueden también incluirse leyendas para cada gráfica que se visualiza. Para ello hay que indicar cada leyenda en el plot correspondiente e indicar con *legend()* dónde se quiere que se sitúe (por defecto busca la mejor posición). 


Por último, podemos también incluir una rejilla con la función grid().

Los [ejemplos con PyPlot](https://matplotlib.org/stable/gallery/index.html#pyplot) son un excelente punto de información para ver cómo y qué se puede hacer con la API de PyPlot de Matplotlib. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-25, 25, 2)

y_1 = x
y_2 = x**2
y_3 = 10*x + 1000
y_4 = -10*x + 1000
```

```python
# Dibujamos líneas
# A cada línea le asignamos un estilo (marcador, tipo de línea, color)  
# y una etiqueta

plt.plot(x,y_1,'.-b', label ='y_1')
plt.plot(x, y_2, 's:r', label = 'y_2') 
plt.plot(x, y_3, '^-.g', label = 'y_3')
plt.plot(x, y_4, '-k', label = 'y_4')

# Visualizamos las etiquetas conforme a una leyenda
# Puede indicarse localización
plt.legend(loc='upper left')


# Indicamos tetos para título y ejes
# El texto puede formatearse
plt.title('Un primer ejemplo', color='red')
plt.xlabel('Valores de x')
plt.ylabel('Valores de y')

# Ejemplo de inserción de texto en figura
plt.text(-20, 600, 'texto en la figura', fontsize=14)
# Podemos formatear texto matemático con comandos en Latex
plt.text(-3, 400, r'$\iint \Phi(x, y)dxdy$', fontsize = 20)

# Incluimos una anotación con una flecha
plt.annotate(r'$y = x^2$', xy=(-10, 175), xytext=(0, 200),
             arrowprops=dict(facecolor='red', shrink=0.05),)

# Añadimos una rejilla
plt.grid(True)

# Visualizamos el gráfico
plt.show()
```

Podemos también representar varios *plots* en cada figura y, también tener varias figuras. Para se utiliza la función *subplot(FCn)*, donde F y C indican, respectivamente, el número de filas y columnas que tendrá la representación, y n es un índice para seleccionar de cuál se trata.

Cuando tenemos varias gráficas en la misma figura, es habitual que las leyendas de ejes y títulos puedan superponerse. Para evitarlo, utilizamos la función *tight_layout()*. Hay una [guía](https://matplotlib.org/stable/tutorials/intermediate/tight_layout_guide.html#) que explica en detalle esta cuestión.

Veamos un ejemplo de una gráfica con una fila y dos columnas.


```python
plt.subplot(121)
plt.plot(x, y_1, 'bo')
plt.title('subplot 121')
plt.xlabel('x')
plt.ylabel('recta')

plt.subplot(122)
plt.plot(x, y_2, 'r--')
plt.title('subplot 122')
plt.xlabel('x')
plt.ylabel('parábola')

plt.tight_layout()
plt.show()
```

Mostremos ahora una figura con dos filas y una columna.

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

#### Representación de imágenes en Matplotlib con Pyplot

Si bien *Matplotlib* cuenta con funciones para leer imágenes, la recomendación es utilizar para ello PILLOW, que Matplotlib sabe representar y que puede convertirse de forma inmediata a un *ndarray*

```python
from PIL import Image
```

```python
file = 'images/chinche_L.png'
print(type(file))
print(id(file))

img  = Image.open(file)
print(type(img))
print(id(img))

print('formato:', img.format)
print('(ancho, alto) =', img.size)
print('modo: ', img.mode)
```

Visualicemos ahora la imagen utilizando la función *imshow()* de Matplotlib. Recuerda que esta imagen ya la visualizamos en el primer *notebook*. 

La función imgshow() devuelve la dirección en memoria del objeto de imagen. Para evitar que se visualice dicha dirección, añadimos ";" al final.

```python
plt.imshow(img);
```

¿Qué ha pasado? Cuando *imshow()* se utiliza para visualizar una imagen en escala de grises no visualiza directamente la imagen, sino un **mapa de color**. Esto es así para aprovechar mejor la capacidad de nuestro sistema visual para discriminar pequeños cambios de intensidad en la imagen. Salvo que se le indique otra cosa, *Matplotlib* utilizará el mapa de color *Viridis*, que es perceptualmente uniforme.

Podemos decirle que utilice un mapa de color gris (*gray*), si queremos obtener el resultado que esperábamos.

En todo caso, adviértase que *imshow()* espera imágenes, no matrices con valores arbitrarios. En tal caso, debemos de normalizarlas para la visualización sea como esperamos. Hay muchas maneras de hacerlo, implicando en dicha normalización al mapa de color también. Lo más sencillo es asegurarnos que representamos:
- Valores entre 0 y 255 cuando la imagen está compuesta de números enteros
- Valores entre 0 y 1 si la imagen es de números reales

En el caso de las imágenes RGB, *imshow()* espera los tres canales con enteros entre 0 y 255

Adviértase también que *imshow()* muestra los ejes.

VEamos cómo hace una representación con el mapa de color gris, y le añadiremos una barra que muestra los valores numéricos de cada nivel de gris.

```python
plt.imshow(img, cmap='gray')
plt.colorbar();
```

```python
img_np = np.asarray(img)
img_lin_h = img_np.copy()
img_lin_v = img_np.copy()

f = 100
c = 240

plt.subplot(221)
img_lin_h[f, :] = 255
plt.imshow(img_lin_h, cmap='gray')


plt.subplot(222)
img_lin_v[:, c] = 255
plt.imshow(img_lin_v, cmap='gray')

plt.subplot(223)
plt.plot(img_np[f,:])

plt.subplot(224)
plt.plot(img_np[:, c])
```

Por último, convirtamos la imagen a números reales y normalicémos entre 0 y 1 para representarla. Ahora utilizaremos el mapa de color *Inferno*, que es también perceptualmente uniforme. 

Puedes ver [aquí](https://matplotlib.org/stable/tutorials/colors/colormaps.html#choosing-colormaps-in-matplotlib) mucho más sobre cómo escoger mapas de color en *Matplotlib*.

```python
img_float = np.asarray(img, dtype = 'float')
print('Tipo de la imagen: ', img_float.dtype)
print('Valor mínimo de imagen float:', img_float.min())
print('Valor máximo de imagen float:', img_float.max())

#Representamos entre 0 y 1
img_float = img_float/img_float.max()
print('Valor mínimo de imagen float normalizada:', img_float.min())
print('Valor máximo de imagen float normalizada:', img_float.max())

plt.imshow(img_float, cmap='inferno')
plt.colorbar();
```
